```python
def partition(self, s: str) -> List[List[str]]:

    n = len(s)
    f = [[True]*n for _ in range(n)]
    # 核心公式就是f[i][j] = f[i+1][j-1]&s[i][j]
    # 所以i要倒序处理，j要顺序处理
    # 在这个地方需要注意，i=n-1这个部分其实是不会被运行到的，因为在j中没法被运行
    for i in range(n-1,-1,-1):
        for j in range(i+1,n):
            f[i][j] = f[i+1][j-1] and s[i]==s[j]

    # 预处理部分完成

    ans =list()
    res =list()
    def dfs(i:int):
        # 如果当前i==n则表明一次循环结束了，添加到res即可
        if i ==n:
            res.append(ans[:])
        # 这个地方一个j会运行多次dfs
        # 然后j会从1到n-1整个部分多次dfs
        for j in range(i+1,n):
            if f[i][j] is True:
                ans.append(s[i:j+1])
                dfs(j+1)
                ans.pop()
    return res
        


```

题目**125**

```python
def isPalindrome(self, s: str) -> bool:
    # 预处理
    # # 将大字符转为小字符，非字符给去除掉
    sstd=str
    for c in s:
        if 65<=ord(c)<=90:
            sstd.append(chr(ord(c)+32))
        if 97<=ord(c)<=122:
            sstd.append(c)

    n=len(sstd)
    left,right=0,n-1
    while left<right:
        if sstd[left] !=sstd[right]:
            return False
        left+=1
        right-=1
    return True
    
```

题目**680**

```python
def validPalindrome(self, s: str) -> bool:
    def checkpoint(left:int,right:int)->bool:
        while left<right:
            if s[left]!=s[right]:
                return False
            left+=1
            right-=1
        return True

    left,right = 0,len(s)-1
    while left <right:
        if s[left]!=s[right]:
            return checkpoint(left+1,right) or checkpoint(left,right-1)
        left+=1
        right-=1
    return True

    

```

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minStack=[math.inf]
            

    def push(self, val: int) -> None:
        self.stack.append(val)
        self.stack.append(min(self.minStack[-1],val))
        
        
    def pop(self) -> None:
        self.stack.pop()
        self.minStack.pop()
        

    def top(self) -> int:
        return self.stack[-1]
        
    def getMin(self) -> int:
        return self.minStack[-1]
        
```

题目**5**  
>这一题要求子串连续的情况下，最长的连续子串长度

```python
def longestPalindrome(self, s: str) -> str:
    # 对于某个回文的子串而言，
    # s[i]==s[j]
    # #f[i][j]=f[i+1][j-1]+2
    # 而s[i]!=s[j]的情况下
    # #f[i][j]=max(f[i-1][j],f[i][j-1])
    n = len(s)
    f = [[True]*n for _ in range(n)]
    for i in range(n-1,-1,-1):
        for j in range(i+1,n):
                f[i][j]=f[i+1][j-1] and s[i]==s[j]
        
    max = 0
    x=0
    y=0
    for i in range(n):
        for j in range(n):
            if f[i][j] and j-i>max:
                max=j-i
                x=i
                y=j
    return s[x:y+1]

```

>题目**516**

>这一题是针对于最长的字串长度，字串本身是可以不连续的部分

```python
def longestPalindrome(self, s: str) -> str:
    # 对于某个回文的子串而言，
    # s[i]==s[j]
    # #f[i][j]=f[i+1][j-1]+2
    # 而s[i]!=s[j]的情况下
    # #f[i][j]=max(f[i-1][j],f[i][j-1])
    n = len(s)
    f= [[0]* n for _ in range(n)]
    for i in range(n):
        f[i][i]=1
    # 预处理完成
    for i in range(n-1,-1,-1):
        for j in range(i+1,n):
            if s[i]==s[j]:
                f[i][j]=f[i+1][j-1]+2
            else:
                f[i][j] = max(f[i][j-1],f[i+1][j])
    i,j = 0,n-1
    left=[]
    right = []
    while i<j:
        if s[i]==s[j]:
            left.append(s[i])
            right.append(s[j])
            i+=1
            j-=1
        else:
            if f[i+1][j]>=f[i][j-1]:
                i+=1
            else:
                j-=1
    return "".join(left+right[::-1])
    
```

题目**136**  
>本题目找只出现过一次的数字  
>用异或运算去处理

```python
# 这种排序后再做的方式有点慢了
def singleNumber(self, nums: List[int]) -> int:
        n =len(nums)
        if n ==1:
            return nums[0]
        nums.sort()
        left=0
        right = 1
        
        while left<n and right<n:
            if nums[left]!=nums[right]:
                return nums[left]

            left+=2
            right+=2
        return nums[left]
        
```

```python
def singleNumber(self, nums: List[int]) -> int:
        n =len(nums)
        res=0
        for n in nums:
            res^=n
        return res
        
```

```python
def majorityElement(self, nums: List[int]) -> int:
    # 用hash表统计结果
    hashtable = dict()
    for i in nums:
        hashtable[i]= hashtable.get(i,0)+1
    return max(hashtable.items(),key=lambda x:x[1])[0]
```
