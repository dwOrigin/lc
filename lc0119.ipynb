```python
def maxSideLength(self, mat: List[List[int]], threshold: int) -> int:
    # 对于本题而言，可以使用和昨天类似的做法去做
    # 依旧是使用前缀和的方式，但是这个只使用行的前缀和
    # 通过最后的遍历处理，来判断当前的值是否满足或者是不满足结果

    # # 当前这种方式求解的时候会超时
    m,n = len(mat),len(mat[0])
    rowSum = [[0]*(n+1) for _ in range(m)]

    # 行的前缀和
    for i in range(m):
        for j in range(n):
            rowSum[i][j+1]=rowSum[i][j]+mat[i][j]
    
    for k in range(min(m,n),0,-1):
        for i in range(k,m+1):
            for j in range(k,n+1):
                if sum(rowSum[r][j]-rowSum[j-k] for r in range(i-k,i))<=threshold:
                    return k
        
    return -1
        
```

```python
#二维的前缀和求解方式
def maxSideLength(self, mat: List[List[int]], threshold: int) -> int:
    m,n = len(mat),len(mat[0])

    # 构造二维前缀和矩阵
    pre = [[0]*(n+1) for _ in range(m+1)]
    
    for i in range(m):
        for j in range(n):
            pre[i+1][j+1] = pre[i+1][j]+pre[i][j+1]-pre[i][j]+mat[i+1][j+1]
    
    # 求子矩阵的和
    def get_sum(x1:int,y1:int,x2:int,y2:int)->int:
        return pre[x2+1][y2+1]-pre[x1][y2+1]-pre[x2+1][y1]+pre[x1][y1]
    
    # 判断是否符合条件
    def check(k :int)->bool:
        for i in range(k-1,m):
            for j in range(k-1,n):
                if get_sum(i-k+1,j-k+1,i,j)<=threshold:
                    return True
        return False
    left,right = 0,min(m,n)
    while left<right:
        mid = (left+right+1)//2
        if check(mid):
            left = mid
        else:
            right = mid-1
            
    return left
    

    
```

```python
# 真前缀树的写法

class Trie:

    def __init__(self):
        self.children=[None]*26
        self.isEnd = False

    def insert(self, word: str) -> None:
        node = self
        for w in word:
            num = w-'a'
            if node.children[num] is None:
                node.children[num] = Trie()
            node = node.children[num]
        node.isEnd = True
    def searchPrefix(self,prefix:str)->"Trie":
        node = self
        for p in prefix:
            ch = p-'a'
            if node.children[ch] is None:
                return None
            node = node.children[ch]
        return node
                    
    def search(self, word: str) -> bool:
        node = self.searchPrefix(word)
        return node is not None and node.isEnd
        
        

    def startsWith(self, prefix: str) -> bool:
        return self.searchPrefix(prefix) is not None
        
```

题目**79**
>本题采用深度优先搜索的方式求解
>先进行对于当前的节点进行判断，
>如果当前节点的值和那个值相同，就进行深度搜索，否则就直接进行下一个
>
>

```python
def exist(self, board: List[List[str]], word: str) -> bool:
    m =len(board)
    n = len(board[0]) 
    visited = [[False]*n for _ in range(m)]
    def dfs(board:List[List[str]],i:int,j:int,word,num:int)->bool:
        
        if board[i][j]!=word[num] or visited[i][j]:
            return False
        
        if num==len(word)-1:
            return True
        
        if board[i][j]==word[num] and visited[i][j]==False:
            visited[i][j]=True
            for x,y in [(0,1),(0,-1),(1,0),(-1,0)]:
                if 0<=i+x<m and 0<=j+y<n and dfs(board,i+x,j+y,word,num+1):
                       return True
            visited[i][j]=False
            
            return False
    for i in range(m):
        for j in range(n):
            if dfs(board,i,j,word,0):
                return True
    return False
            

```
