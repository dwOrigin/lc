```python
# 62
def uniquePaths(self, m: int, n: int) -> int:
    res=[[1 for _ in range(n)] for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if i==0 or j ==0:
                continue
            res[i][j]=res[i-1][j]+res[i][j-1]
    return res[m-1][n-1]
```

```python
#34
# 这个题目转变为了，找比当前值小一个的数字和找比当前值大一个的数字
def searchRange(self, nums: List[int], target: int) -> List[int]:
    def binSearch(nums:List[int],target:int):
            l= 0
            r = len(nums)-1
            while(l<=r):
                mid = (l+r)//2
                if nums[mid]<target:
                    l = mid+1
                if nums[mid]>=target:
                    r = mid-1
            return l
    res = [-1,-1]
    le = binSearch(nums,target)
    if le == len(nums) or nums[le]!= target:
        return res
    r = binSearch(nums,target+1)-1
    res[0]=le
    res[1]=r
    return res

```

```python
#64
def minPathSum(self, grid: List[List[int]]) -> int:
    m = len(grid)    # 行数
    n = len(grid[0]) # 列数
    
    for i in range(m):
        for j in range(n):
            if i == 0 and j == 0:
                continue # 起点不用变
            elif i == 0:
                grid[i][j] += grid[i][j-1] # 第一行，只能从左边加过来
            elif j == 0:
                grid[i][j] += grid[i-1][j] # 第一列，只能从上边加过来
            else:
                grid[i][j] += min(grid[i-1][j], grid[i][j-1]) # 取左边和上边的较小者
                
    return grid[m-1][n-1]
```

```python
#1143
# 使用和较短的那个相同的长度大小的数组作为dp
def longestCommonSubsequence(self, text1: str, text2: str) -> int:
    m=len(text1)
    n= len(text2)
    dp = [[0]*(m+1) for _ in range(n+1)]
    for i in range(1,m+1):
        for j in range(1,n+1):
            if text1[i-1]==text2[j-1]:
                dp[i][j] = dp[i-1][j-1]+1
            else:
                dp[i][j]=max(dp[i-1][j],dp[i][j-1])

    return dp[m][n] 
    


```

```python
#48
#对整个图像的旋转进行数学公式的计算，然后直接求解即可
# matrix[i][j]最终变换到的位置是matrix[j][n-1-i]
# 先进行上下对称变换
# 再进行轴对称中心变换
def rotate(self, matrix: List[List[int]]) -> None:
        """
        Do not return anything, modify matrix in-place instead.
        """
        m = len(matrix)
        n = len(matrix[0])
        for i in range(m//2):
            for j in range(n):
                  matrix[i][j],matrix[n-i-1][j]=matrix[n-i-1][j],matrix[i][j]
        
        for i in range(m):
              for j in range(i,n):
                    matrix[i][j],matrix[j][i]=matrix[j][i],matrix[i][j]
```

```python
#240
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m = len(matrix)
    n = len(matrix[0])
    for i in range(m):
        l=0
        r= n-1
        while l<=r:
            mid = (l+r)//2
            if matrix[i][mid] == target:
                return True
            if matrix[i][mid]<target:
                l = mid+1
            else:
                r = mid-1
    return False

```

```python
#17电话号码

def letterCombinations(self, digits: str) -> List[str]:
    phoneMap={
        "2":"abc",
        "3":"def",
        "4":"ghi",
        "5":"jkl",
        "6":"mno",
        "7":"pqrs",
        "8":"tuv",
        "9":"wxyz"
    }
    res = list()
    combination = list()
    def back(index:int):
        if index ==len(digits):
            res.append("".join(combination))
        else:
            for letter in phoneMap[digits[index]]:
                combination.append(letter)
                back(index+1)
                combination.pop()

    back(0)
    return res

```

```python
#39组合总和
# 普通的回溯
def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
    candidates.sort()
    res = list()
    onetime = list()
    def backreverse(start:int,curr:int):
        if curr == 0:
            res.append(list(onetime))
        else:
            for i in range(start,len(candidates)):
                if curr-candidates[i]<0:
                    break
                else:
                    onetime.append(candidates[i])
                    backreverse(i,curr-candidates[i])
                    onetime.pop()
    
    backreverse(0,target)
    return res
```

```python
#22
# 括号生成
def generateParenthesis(self, n: int) -> List[str]:
    res =list()
    curr =''
    def backtracking(curr:list,left:int,right:int):
        if left==0 and right ==0:
            res.append(curr)
        if right < left:
            return
        if left>0:
            backtracking(curr+"(",left-1,right)
        if right>0:
            backtracking(curr+")",left,right-1)

    backtracking(curr,n,n)
    return res


```
