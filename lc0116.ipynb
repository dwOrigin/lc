> 这个题目的整体思路是对的，就是为了求得两条边上的各自最大最合适的值
>被注释掉的部分是没有进行append 1 m的时候做的，这种方式需要补来补去的
>不符合算法规范，而在hFences上进行添加这两个部分后，可以直接求也不用补
>
>题目**2975**

```python
def maximizeSquareArea(self, m: int, n: int, hFences: List[int], vFences: List[int]) -> int:
# 此时可以假想为没有所谓的栏杆，
# 先计算出来，不同的vFence标记点下面的长度是多少，然后存储起来
# 然后再去计算不同的hFence标记点下面的长度是多少，然后存储起来
# 遍历第二个hashtable，看看第一个里面是否有对应的数据，
# 如果有的话，那么就是要求解的值。如果最后两者都没有相同的值
# 就返回-1的结果

    vHash = set()
    hHash = set()
    # hHash.update(x - 1 for x in hFences)
    hFences.append(1)
    hFences.append(m)
    hFences.sort()
    for i in range (len(hFences)-1):
        for j in range(i+1,len(hFences)):
            hHash.add(hFences[j]-hFences[i])
        # hHash.add(m-hFences[i])
    # hHash.add(m-hFences[-1])
    # hHash.add(m-1)
    vFences.append(1)
    vFences.append(n)
    vFences.sort()
    # vHash.update(x - 1 for x in vFences)
    for i in range (len(vFences)-1):
        for j in range(i+1,len(vFences)):
            vHash.add(vFences[j]-vFences[i])
        # vHash.add(n-vFences[i])     
    # vHash.add(n-vFences[-1])
    # vHash.add(n-1)
    lst = sorted(hHash, reverse=True)    
    for num in lst:
        if num in vHash:
            return num*num    
    return -1
        
```

>题目**138**
>采用交叉节点的思想这一点很好，避免了使用hashtable
>
>然后可以进一步提升对于链表的理解

```python
from typing import Optional
```

```python
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
```

```python
class Solution:
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        # 采用类似于交错节点的方式去做，
        # 将新节点插入到原节点的后面去
        # 然后依次遍历新节点中的属性
        # 将新节点中的相关属性值，用新节点赋予
        if head is None:
            return None
        current = head

        while current is not None:
            current.next = Node(current.val,current.next)
            current = current.next.next

        # 交叉节点完成，然后对于random进行更新
        current = head
        while current is not None:
            if current.random is not None:
                current.next.random = current.random.next
            
            current = current.next.next

        # random也更新好了，然后需要进行修剪了
        newHead = head.next
        copy = newHead
        head = head.next.next
        while head is not None:
            copy.next = head.next
            copy = copy.next
            head = head.next.next

        return newHead
```

>题目**148**
>
>采用归并排序的思想来做

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    if head is None or head.next is None:
         return head
    
    head2= self.middleNode(head)
    head = self.sortList(head)
    head2 = self.sortList(head2)

    return self.twoSortedList(head,head2)
    
def middleNode(self,head: Optional[ListNode])->Optional[ListNode]:
        slow = head
        fast = head
        pre = slow
        while fast and fast.next:
            pre = slow
            slow = slow.next
            fast = fast.next.next
        pre.next = None
        return slow

def twoSortedList(self,list1: Optional[ListNode],list2:Optional[ListNode])->Optional[ListNode]:
        
        curr = dummy = ListNode()
        while list1 and list2 :
            if list1.val <list2.val:
                curr.next = list1
                list1 = list1.next
            else:
                curr.next = list2
                list2 = list2.next
            curr = curr.next
        curr.next = list1 if list1 else list2

        return dummy.next    
```
