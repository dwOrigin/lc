题目**3047**
>这个题目通过有点太顺利了，直接采用的是构建面积的思想
>两者的交叉部位中取交集即可，然后即可得到相关的正方形的体积了

```python
#本题和昨天那个题目有点像
# 任意两个正方形之间的相互交集面积而言，
# 左下方向max{x,y},右上方向min{x,y}
# 如果左下的x等于大于右上的x
# 或者是左上的y大于等于右上的y
# 那么二者之间的值都是为0的，否则就是有值的
# 对于最大的正方形的面积来说，那就是取min{长，宽}
# ---代码实现---
# 可以两个图形依次比较，这样的话会比较快，节约内存空间
def largestSquareArea(self, bottomLeft: List[List[int]], topRight: List[List[int]]) -> int:
    outer=len(bottomLeft)
    res = 0
    for i in range(outer-1):
        for j in range(i+1,outer):
            blx = max(bottomLeft[i][0],bottomLeft[j][0])
            bly = max(bottomLeft[i][1],bottomLeft[j][1])
            trX = min(topRight[i][0],topRight[j][0])
            trY = min(topRight[i][1],topRight[j][1])
            if blx>=trX or bly>=trY:
                continue
            res = max(res,min(trX-blx,trY-bly)**2)

    return res
```

```python
# use typing List for type annotations
from typing import List
```

```python
def numIslands(self, grid: List[List[str]]) -> int:
    def dfs(grid: List[List[str]],i,j):
        if i<0 or j<0 or i>=len(grid) or j>=len(grid[0]):
            return
        if grid[i][j]=='1':
            grid[i][j]='0'
        dfs(grid,i+1,j)
        dfs(grid,i,j+1)
        dfs(grid,i-1,j)
        dfs(grid,i,j-1)
    # 本题进行深度遍历，进行一次深度遍历就代表有一个岛屿
    # 用for循环去检测全部的grid网格，没有遍历过的就给遍历上
    counter=0
    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j]=='1':
                dfs(grid,i,j)
                counter+=1
    return counter
```

```python
# 如果是广度优先遍历的方式呢，应该怎么处理呢
def numIslands(self, grid: List[List[str]]) -> int:
    num_islands = 0
    for r in range(len(grid)):
        for c in range(len(grid[0])):
            if grid[r][c] =='1':
                num_islands+=1
                grid[r][c]='0'
                neighbors = collections.deque([(r,c)])
                while neighbors:
                    row,col = neighbors.popleft()
                    for x,y in [(row-1,col),(row+1,col),(row,col-1),(row,col+1)]:
                        if 0<=x<len(grid) and 0<=y<len(grid[0]) and grid[x][y]=='1':
                            grid[x][y]='0'
                            neighbors.append((x,y))

    return num_islands
```
