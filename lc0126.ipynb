```python
from typing import List
```

```python
def minimumAbsDifference(self, arr: List[int]) -> List[List[int]]:
    arr.sort()
    # 创建一个值专门用来存储当前最小的值，
    # 移动数组进行比较，如果是等于这个值，存起来，小于这个值，数组清空重新存储
    minNum = float("inf")
    res = list()
    for i in range(1,len(arr),1):
        temp = arr[i]-arr[i-1]
        if temp<minNum:
            res.clear()
            res.append(list(arr[i-1],arr[i]))
            minNum = temp
    return res
```

```python
def longestValidParentheses(self, s: str) -> int:
    # 构建一个空的栈，只用来存储（，
    # 构建一个等价的数组,来判断连续的个数
    # 最后
    if not s:
        return 0
    st = list()
    res = 0
    loc = [0] *len(s)
    cur=0

    for i in range(len(s)):
        if s[i] =="(":
            st.append(i)
        elif st and s[i] == ")":
            j=st.pop()
            loc[i]=1
            loc[j] = 1
    
    for num in loc:
        if num:
            cur+=1
        else:
            res = max(cur,res)
            cur=0
    res = max(cur,res)
    return res
```

>利用官方的出栈策略,在遍历的时候将当前的(的下标给记录下来  

```python
def longestValidParentheses(self, s: str) -> int:
    st = list()
    res = 0
    st.append(-1)
    for i,ss in enumerate(s):
        if ss == "(":
            st.append(i)
        elif st:
            st.pop()
            res = max(i-st[-1],res)
            
    return res 
```

>这种方式是使用了Counter这个结构  
>减少了统计当前的t以及s子串的个数  
>不是完全的从基础思想出发的

```python
def minWindow(self,s:str,t:str) ->str:
    cnt_s = Counter()
    cnt_t = Counter(t)

    ans_left,ans_right = -1,len(s)
    left = 0
    for right,c in enumerate(s):
        cnt_s[c]+=1
        while cnt_s>=cnt_t:
            if right-left<ans_right-ans_left:
                ans_left,ans_right= left,right

            cnt_s[s[left]]-=1
            left+=1

    return "" if ans_left<0 else s[ans_left,ans_right+1]
```

```python
from collections import defaultdict

def minWindow(self,s:str,t:str) ->str:
    # 用hash表cnt表示当前子串中对应的某个字母和t中相应字母相比，少的个数
    # 用less表示当前子串整体比t少的字母的个数有几个

    cnt = defaultdict(int)
    for i in t:
        cnt[i]+=1
    less = len(cnt)

    ans_left = -1
    ans_right = len(s)
    left = 0
    for right,c in enumerate(s):
        cnt[c]-=1

        if cnt[c]==0:
            less-=1
        while less == 0:
            if right - left<ans_right-ans_left:
                ans_left,ans_right = left,right
            x = s[left]    
            
            if cnt[x] == 0:
                less +=1

            cnt[x] += 1
            left+=1
    
    return "" if ans_left <0 else s[ans_left:ans_right+1]
        
```

题目**4**
>对于本题，采用左右两个最大堆和最小堆去实现呢，刚开始的时候  
>先将num1给全部分两部分单独给对应的left的大堆以及right的小堆  
>保证左侧的值全部小于右边的值，然后单独添加num2的值进入到左右两个堆  
>当len(left)==len(right)的时候，添加到左边
>当len(left)>len(right)的时候，添加到左边，然后将左边的最大的值给右边


```python
import heapq
```

```python
def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
    h = nums1 + nums2
    heapq.heapify(h)
    n = len(h)
    if n % 2 == 1:
        for _ in range(n//2):
            heapq.heappop(h)
        return heapq.heappop(h)
    else:
        for _ in range(n//2 - 1):
            heapq.heappop(h)        
        a = heapq.heappop(h)
        b = heapq.heappop(h)
        return (a + b) / 2
            
```

>按照规范的做法去实现这项内容
>将左右两侧根据有序的定义，分为两段，
>左端两段的值都小于右端两边的值即可
>

```python
def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1)>len(nums2):
            nums1,nums2 = nums2,nums1

        m,n=len(nums1),len(nums2)
        nums1 = [-inf]+nums1+[inf]
        nums2 = [-inf]+nums2+[inf]
        # nums1中有i个数在第一组
        # 那么nums2中有（m+n+1）//2 -i个数在第二组
        i,j = 0,(m+n+1)//2

        while True:
            if nums1[i]<=nums2[j+1] and nums2[j]<=nums1[i+1]:
                max1 = max(nums1[i],nums2[j])
                min2 = min(nums1[i+1],nums2[j+1])
                return max1 if (m+n)%2 else (max1+min2)/2
            i+=1
            j-=1
                    
```
