```python
from typing import List
from  heapq import *
```

```python
def minCost(self, n: int, edges: List[List[int]]) -> int:
    #构建邻接表 
    # 构建距离
    g = [[] for _ in range(n)]

    for x,y,wt in edges:
        g[x].append((y,wt))
        g[y].append((x,wt*2))
    inf = float("inf")
    dis = [inf]*n

    dis[0] = 0
    h = [(0,0)] #(起点0到节点的距离，节点)

    while h:
        dis_x,x = heappop(h)
        if dis_x >dis[x]:#说明此时不需要更新dis[x]
            continue
        if x ==n-1: 
            #为什么这个地方就可以返回了呢?
            # 采用的是heappop，每次从堆里拿出的点，一定是「当前所有未确定点里距离最小的」
            # 如果能pop n-1节点的话，说明内容已经全部被完成了
            # 所以这个时候已经可以去返回节点了
            return dis[x]
        
        for y,wt in g[x]:
            new_dis_y = dis_x+wt
            if new_dis_y<dis[y]:
                dis[y] = new_dis_y
                heappush(h,(new_dis_y,y))
    return -1
```

```python
def networkDelayTime(self, times: List[List[int]], n: int, k: int) -> int:
    # 构建邻接表
    # 构建记录k到其他的节点的时间
    # 遍历
    g = [[] for _ in range(n)]
    inf = float("inf")
    dis = [inf] *n
    h = [(0,k-1)]
    dis[k-1]=0
   

    for x,y,wt in times:
        g[x-1].append((y-1,wt))

    while h:
        dist,x = heappop(h)
        if dist>dis[x]:
            continue

        for y,d in g[x]:
            new_dis_k = dist +d
            if dis[y]>new_dis_k:
                dis[y]=new_dis_k
                heappush(h,(new_dis_k,y))
    return -1 if max(dis)==inf else max(dis)
    
```

题目**72**
>对于这道题目而言，采用动态规划  
>将原本的增、删、替三个对于a的操作  
>修改为了A增、B增、A替的操作  
>这样避免了对于删除操作的把控，更方便写一些  
>然后就是需要进行处理，
>分为d[i-1][j] d[i][j-1] d[i-1][j-1]
>假设当前这三个都已经处理好了，那么对于d[i][j]而言
>d[i-1][j]只能进行B的add操作，为d[i-1][j]+1次操作
>d[i][j-1]只能进行A的add操作，为d[i][j-1]+1次操作
>d[i-1][j-1]可以进行A的adjust操作或者是B的adjust操作即d[i-1][j-1]+1，或者i和j本身就相同直接就是d[i-1][j-1]  
>运行的过程中需要注意 i，j本身是存在的，只是调整运行的步骤让它们相通，并不是走一步，然后下一步随便加两个i，j元素

```python
def minDistance(self, word1: str, word2: str) -> int:
    m,n = len(word1),len(word2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for i in range(m):
        dp[i+1][0] = i+1
    for j in range(n):
        dp[0][j+1] = j+1

    for i in range(1,m+1):
        for j in range(1,n+1):
            if word1[i] == word2[j]:
                dp[i][j] = min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])+1
            else:
                dp[i][j] = min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1]-1)+1
    return dp[m][n]
     
```
