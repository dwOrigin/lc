题目**1895**

```python
from typing import List
```

```python
def largestMagicSquare(self, grid: List[List[int]]) -> int:
    # 依旧采用全部遍历的思想
    # 对于每个需要遍历的内容进行处理幻方判断处理
    # 先对于行求一次前缀和  rowSum[i][j+1]是对i行的[0~j]元素的前缀和
    # #rowSum中，对于row是[0,m-1]
    # 再对于列求一次前缀和  colSum[i+1][j]是对j列的[0,i]元素的前缀和
    # #colSum中，对于col是[0,n-1]
    # 再对主对角线求一次前缀和 diagSum[i+1][j+1]左上到右下
    # 再对副对角线求一次前缀和 antiSum[i+1][j+1] 右下到左上
    m,n = len(grid),len(grid[0])
    rowSum = [[0]*(n+1) for _ in range(m)]
    colSum = [[0]*(n) for _ in range(m+1)]
    diagSum = [[0]*(n+1) for _ in range(m+1)]
    antiSum = [[0]*(n+1) for _ in range(m+1)]
    # 对其进行初始化
    for i,row in enumerate(grid):
        for j,x in enumerate(row):
            rowSum[i][j+1] = rowSum[i][j]+x
            colSum[i+1][j] = colSum[i][j]+x
            diagSum[i+1][j+1] = diagSum[i][j]+x
            antiSum[i+1][j] = antiSum[i][j+1]+x
    
    # 从反方向开始算k的最大值
    for k in range(min(m,n),0,-1):
        for i in range(k,m+1):
            for j in range(k,n+1):
                num = diagSum[i][j]-diagSum[i-k][j-k]
                # 反对角线的和
                # 行的和
                # 列的和
                if antiSum[i][j]-antiSum[i-k][j-k]==num \
                and all(rowSum[r][j]-rowSum[r][j-k]==num for r in range(i-k,i)) \
                    and all(colSum[i][r]-colSum[i-k][r]==num for r in range(j-k,j)):
                    return k
```

题目**207**

```python
import collections
```

```python
def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
    # 对于这一个题目而言，可以使用的是
    # 构建拓扑排序也就是有向无环图
    # 构建一个栈，然后将遍历的内容压入到栈中，
    # 每出栈一次，则认为完成了一门课程，等到全部出栈后
    # 利用广度优先搜索的方式去求解
    edges = collections.defaultdict(list)
    indegree = [0]*numCourses

    for info in prerequisites:
        edges[info[1]].append(info[0])
        indegree[info[0]]+=1

    q = collections.deque([u for u in range(numCourses) if indegree[u]==0])
    visited = 0
    while q:
        visited+=1
        u = q.popleft()
        for v in edges[u]:
            indegree[v]-=1
            if indegree[v]==0:
                q.append(v)
    
    return visited == numCourses
```
